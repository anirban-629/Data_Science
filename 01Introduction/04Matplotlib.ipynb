### Axes, Labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Data
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,52.5]
min_t=[43,42,40,35,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

#Label & Legend
plt.plot(days,max_t,label="MAX")
plt.plot(days,min_t,label="MIN")
plt.plot(days,avg_t,label="AVG")

# plt.legend() # By default loc="best" (Means= depending on chart matplotlib decides the postion)
# plt.legend(loc="lower left",shadow=True)
# plt.legend(loc=3,shadow=True) #loc="lower left" In case of loc lower left we can use the code 3
plt.legend(loc=0,shadow=True,fontsize="small")
```

```python
plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

# Grid
plt.plot(days,max_t,label="MAX")
plt.plot(days,min_t,label="MIN")
plt.plot(days,avg_t,label="AVG")

plt.legend(loc=0,shadow=True,fontsize="small")
plt.grid()
```

- #### Introducing the new way of Representing Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Data
company=["Google","Amazon","Microsoft","Apple"]
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)            # Giving the names of the companies 
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
# Horizontal Chart

plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")

plt.yticks(xpos,company)            # Giving the names of the companies 
plt.xlabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
# Data
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.title("Blood Sugar Analysis")
plt.ylabel("Total no. of Paitents")
plt.xlabel("Number of Paitents")

plt.hist(
    [blood_sugar_men,blood_sugar_women],
    rwidth=0.99,
    bins=[80,100,125,150],
    color=['g','orange'],
    label=['Men','Women']
)
plt.legend()
```

```python
plt.title("Blood Sugar Analysis")
plt.xlabel("Total no. of Paitents")
plt.ylabel("Sugar range")

plt.hist(
    [blood_sugar_men,blood_sugar_women],
    rwidth=0.99,
    bins=[80,100,125,150],
    color=['g','orange'],
    label=['Men','Women'],
    orientation='horizontal'
)
plt.legend()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Data
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
# plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show() # Only gives the pie chart
```

```python
plt.pie(
    exp_vals,
    labels=exp_labels,
    radius=1.5,
    autopct='%0.2f%%', # Selects the decimal point
    explode=[0,0,0,0,1], # Detouchs Single pie along with the amount of detouch like (1,2..)
)
plt.show()
```

```python
plt.pie(
    exp_vals,
    labels=exp_labels,
    radius=1.5,
    autopct='%0.2f%%', 
    explode=[0,0,0,0,1], 
    startangle=45 # By default start from 0 and with L to R order
)
plt.show()
```

- #### Saving the piechart

```python
plt.pie(
    exp_vals,
    labels=exp_labels,
    radius=1.5,
    autopct='%0.2f%%', 
    explode=[0,0,0,0,1], 
    startangle=45
)

plt.savefig(
    'firstPieChart.png',
    bbox_inches='tight',
    pad_inches=2,
#     transparent=True
)
```
